```python
dataset=pd.read_csv('Big_mart.csv')
dataset
```

```python
a=dataset['Item_Fat_Content'].replace(['Regular','Low Fat','reg','low fat'],['R','LF','R','LF'],inplace=True)
a=dataset['Outlet_Size'].fillna(value='medium',inplace=True)
```

```python
a=dataset['Outlet_Size'].replace(['Medium','medium','High','small','Small'],['Medium','Medium','High','Small','Small'],inplace=True)
a=dataset.fillna(dataset.mean())
a
```

```python
S=a.loc[:,['Item_Visibility','Item_Weight','Item_MRP','Item_Outlet_Sales','Item_Fat_Content']]
S
```

```python
S1=S.head(100)
S1
```

```python
x=S1.iloc[:,:-1].values
y=S1.iloc[:,4].values
```

```python
print(pd.DataFrame(x))
print(pd.DataFrame(y))
```

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
print(pd.DataFrame({'Actual':list(y_test),'Predicted':list(y_pred)}))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
